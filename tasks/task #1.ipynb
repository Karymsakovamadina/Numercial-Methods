```python
a = 3.05
b = 3

error = abs(a - b)
delta = error / abs(a)

print(f'Точное значение: {a}')
print(f'Приближенное значение: {b}')
print(f'Абсолютная погрешность: {error}:')
print(f'Относительная погрешность: {delta}')
print(f'Относительная погрешность (%): {delta * 100}')
```

```python
def f(x):
    return x ** 0.5


def f_(x):
    return 1 / (2 / (x ** 0.5))

x = 1.05
error = 0.01


# Вычисление относительной погрешности
delta = abs(f_(x)) * error / f(x)

print("Абсолютная погрешность:", error)
print("Относительная погрешность:", delta)
print(f'Относительная погрешность (%): {delta * 100}')
```
